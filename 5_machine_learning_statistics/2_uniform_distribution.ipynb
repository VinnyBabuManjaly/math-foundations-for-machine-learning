# Uniform distribution
The uniform distribution is a fundamental probability distribution in statistics and machine learning. It's used to model scenarios where all outcomes within a given interval are equally likely—making it one of the simplest but most important distributions.


The uniform distribution can be **discrete** or **continuous**:

- Discrete Uniform Distribution: All n outcomes between a and b are equally probable. For example, rolling a fair die gives values from 1 to 6, each with probability 1/6.

- Continuous Uniform Distribution: Any value in the interval [a, b] is possible, with equal probability density everywhere in that range. For instance, a perfectly random spinner between 0 and 360 degrees.

Mathematical Definitions:

**Discrete**: Each value x in {a, a+1, ..., b} has P(X=x) = 1/n, where n = b−a+1.

**Continuous** (PDF): f(x) = 1/(b−a) for a ≤ x ≤ b

**Expectation** (Mean): μ = (a+b)/2

**Variance**:

- Discrete: (n²−1)/12

- Continuous: (b−a)²/12

```python
# Continuous Uniform: Generate and Visualize

import numpy as np
import matplotlib.pyplot as plt

a, b = 0, 10
size = 10000

# generates uniform values in [a,b)
samples = np.random.uniform(a, b, size)
plt.hist(samples, bins=30, density=True)
plt.title("Histogram: Continuous Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# Discrete Uniform Example

import random

a, b = 1, 6  # Die roll
rolls = [random.randint(a, b) for _ in range(10000)]
# Count occurrences for each value
counts = {x: rolls.count(x) for x in range(a, b+1)}
print(counts)
```

```python
# Using SciPy (Built-in)
# Continuous Uniform

from scipy.stats import uniform

a, b = 0, 10
# uniform(loc, scale) creates uniform distribution from loc (a) to loc + scale (b).
dist = uniform(loc=a, scale=(b - a))  # Uniform over [a, b]

# Generate random values
samples = dist.rvs(size=10000)
# Probability Density Function at x = 5
pdf_5 = dist.pdf(5)
# Cumulative Distribution Function at x = 5
cdf_5 = dist.cdf(5)
```

**Detailed example of uniform distribution implementation using Scipy**

```python
from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt

# loc is left edge of distribution, i.e a
# scale is the width, i.e b-a
# loc and scale values change with respect to each type of continuous distributions
U = uniform(loc=10, scale=5)
```

```python
upoints = np.linspace(7, 18, 200)

ppoints = U.pdf(upoints)

plt.plot(upoints, ppoints)
```

```python
# Mean
U.mean()
```

```python
# Variance
U.var()
```

```python
# Standard deviation
U.std()
```

```python
# Random Variables
U.rvs()
```

```python
# Random Variables
U.rvs(size=3)
```

```python
plt.hist(U.rvs(size=100000))
plt.plot(upoints, ppoints)
```

```python
# Probability of 10
U.pdf(10)
```

```python
# Probability of 11 or 12
U.pdf(11) + U.pdf(12)
```

```python
# Cumulative Probability of x < 11
U.cdf(11)
```

```python
# Cumulative Probability of x > 10
1 - U.cdf(10)
```

```python
# Cumulative Probability between 10 and 11
U.cdf(11) - U.cdf(10)
```

```python
dist1 = uniform(loc=5, scale=10)


sample_means = []

for size in range(1, 501):

    sample = dist1.rvs(size=size, random_state=22)

    sample_means.append(np.mean(sample))


# Answer check
print(type(sample_means))
print(sample_means[:5], '\n', sample_means[-5:])
```

```python

samples_30_or_more = sample_means[29:]
samples_mean = np.mean(samples_30_or_more)
samples_std = np.std(samples_30_or_more)

plt.hist(samples_30_or_more, edgecolor = 'black', alpha = 0.3)
plt.title('Distribution of Sample Means');

# Answer check
print(type(samples_30_or_more))
print(samples_mean)
print(samples_std)
```

```python
# Discrete Uniform

from scipy.stats import randint
a, b = 1, 6
dist = randint(a, b+1)  # Uniform over integers [1, 6]

# Generate random rolls
rolls = dist.rvs(size=10000)
# Probabilities (PMF)
pmf = dist.pmf([1, 2, 3, 4, 5, 6])
# CDF at 3 (P(X ≤ 3))
cdf_3 = dist.cdf(3)
```

| Distribution | Manual/Numpy               | SciPy Built-In            | Key Methods   |
| ------------ | -------------------------- | ------------------------- | ------------- |
| Continuous   | np.random.uniform(a, b, N) | uniform(loc=a, scale=b-a) | rvs, pdf, cdf |
| Discrete     | random.randint(a, b)       | randint(a, b+1)           | rvs, pmf, cdf |
